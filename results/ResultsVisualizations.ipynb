```python
# @title Configure directories
from pathlib import Path

# @markdown Path to results.csv
results = '/content/gdrive/MyDrive/CSC 570 - GAMES/CSC570 - Project/results.csv' # @param {type: 'string'}
```

```python
import pandas as pd
```

```python
df = pd.read_csv(results, index_col=0)
df.head(100)
```

```python
df_mirror = df.copy()
df_mirror.head(3)
for i in df_mirror.index:
    df_mirror['a1'][i], df_mirror['a2'][i] = df_mirror['a2'][i], df_mirror['a1'][i]
    df_mirror['d1'][i], df_mirror['d2'][i] = df_mirror['d2'][i], df_mirror['d1'][i]
    df_mirror['wins'][i], df_mirror['loses'][i] = df_mirror['loses'][i], df_mirror['wins'][i]
    df_mirror['wr'][i] = 1 - df_mirror['wr'][i]

df_mirror.head(10)
```

```python
df_com = df.append(df_mirror, ignore_index=True)
```

```python
deck_matchups = df_com.groupby(by=['d1','d2']).mean()
deck_matchups = deck_matchups.drop(columns=['wins', 'loses', 'ties'])
deck_matchups.head(9)
```

```python
agent_matchups = df_com.groupby(by=['a1','a2']).mean()
agent_matchups = agent_matchups.drop(columns=['wins', 'loses', 'ties'])
agent_matchups.head(100)
```

```python
agentdecks_matchups = df_com.groupby(by=['a1','d1','d2']).mean()
agentdecks_matchups = agentdecks_matchups.drop(columns=['wins', 'loses', 'ties'])
agentdecks_matchups.head(100)
```
